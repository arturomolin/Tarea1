```python
import random  # se importa la libreria random la cual nos permitira generar números aleatorios

def solicitar_matriz(tamaño):
    #se pide al usuario los valores de una matriz cuadrada 50X50
    matriz = []  # pone la matriz vacía
    print(f"Ingresa los valores de la matriz {tamaño}x{tamaño}, fila por fila seárando los valores por un espacio:")
    for i in range(tamaño):  # se itera sobre las filas de la matriz
        fila = list(map(int, input(f"Fila {i + 1}: ").split()))  # se lee una fila de numeros
        while len(fila) != tamaño:  # se verifica que la fila tenga el número correcto de elementos en este caso 50
            print(f"La fila debe tener {tamaño} elementos. Porfavor intentalo nuevamente.")
            fila = list(map(int, input(f"Fila {i + 1}: ").split()))
        matriz.append(fila)  # se agrega la fila a la matriz
    return matriz

def generar_matriz(tamaño):
    #se genera una matriz cuadrada de 50 X 50

    return [[random.randint(1, 100) for _ in range(tamaño)] for _ in range(tamaño)]  # Crea la matriz con valores aleatorios de 1 a 100

def imprimir_matriz(matriz):
    for fila in matriz:  # se itera sobre cada fila de la matriz
        print(" ".join(map(str, fila)))  # se convierte la fila en una cadena y la imprime
    print()  # se agrega una línea en blanco por estetica

def multiplicar_bloques(matriz_a, matriz_b, tamaño, bloque):
    # se multiplican dos matrices usando el metodo de multiplicacion por bloques.
    resultado = [[0] * tamaño for _ in range(tamaño)]
    for i in range(0, tamaño, bloque):  # se itera sobre los bloques de filas
        for j in range(0, tamaño, bloque):  # se itera sobre los bloques de columnas
            for k in range(0, tamaño, bloque):  # se itera sobre los bloques internos
                for x in range(i, min(i + bloque, tamaño)):  # se itera dentro del bloque de filas
                    for y in range(j, min(j + bloque, tamaño)):  # se itera dentro del bloque de columnas
                        suma = 0  # se define una variable para almacenar la suma parcial
                        for z in range(k, min(k + bloque, tamaño)):  # se itera sobre los elementos a multiplicar
                            suma += matriz_a[x][z] * matriz_b[z][y]  # se hace la multiplicación y guarda el resultado
                        resultado[x][y] += suma  # se suma el resultado parcial a la matriz final
    return resultado

def potencia_matriz(matriz, exponente, tamaño, bloque):
    # se calcula la potencia de una matriz usando multiplicacion por bloques.

    resultado = [[1 if i == j else 0 for j in range(tamaño)] for i in range(tamaño)]  #matriz identidad
    base = matriz  # se copia la matriz base

    while exponente > 0:  # se itera hasta que el exponente sea 0
        if exponente % 2 == 1:  # Si el exponente es impar se multiplica por la base
            resultado = multiplicar_bloques(resultado, base, tamaño, bloque)
        base = multiplicar_bloques(base, base, tamaño, bloque)  # se eleva la base al cuadrado
        exponente //= 2  # se reduce el exponente a la mitad

    return resultado

# se piden los datos al usuario
tamaño = 50  # aqui se queda el tamaño fijo aunque se podria cambiar si se decea
bloque = int(input("Ingresa el tamaño del bloque para la multiplicación: "))  # se pide el tamaño del bloque
exponente = int(input("Ingresa el exponente al que desea elevar la matriz: "))  # se pide el exponente

# se eleige como obtener la matriz de forma autonmatica o manual
opcion = input("¿Desea ingresar la matriz manualmente (M) o generarla aleatoriamente (A)? ").strip().upper()
if opcion == "M":
    matriz = solicitar_matriz(tamaño)
else:
    matriz = generar_matriz(tamaño)

print("Matriz original:")  # imprime la matriz original
timprimir_matriz(matriz)

# se calcula la potencia de la matriz
resultado = potencia_matriz(matriz, exponente, tamaño, bloque)

print(f"Matriz elevada a la potencia {exponente}:")  # Muestra el resultado final de la matriz
imprimir_matriz(resultado)


```
