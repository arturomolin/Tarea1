```python
#Jorge Israel
import math  # se importa la biblioteca matematica para poder definir las  funciones

# se definen a las variables
def newton(f, df, x0, tol=1e-5, max_iter=100):
    for i in range(max_iter):
        fx = f(x0)      # se evalua la funcion
        dfx = df(x0)    # se evalua la derivada
        if dfx == 0:    # se verifica si la derivada es cero para asi evitar la division por cero
            print("la derivada es cero por ende no se puede continuar :((.")
            return None
        x1 = x0 - fx / dfx  # se define la formula del metodo de Newton
        if abs(x1 - x0) < tol:  # Si la diferencia entre iteraciones es menor que la tolerancia pedida se converge
            return x1
        x0 = x1  # se actualiza el valor de x para la siguiente iteracion
    print("No se convergio en", max_iter, "iteraciones.")  # Se imprime el mensaje si se excede el numero maximo de iteraciones
    return None

# se defien el inciso a: f(x) = e^x + 2^(-x) + 2cos(x) - 6
def fa(x):
    return math.exp(x) + 2**(-x) + 2*math.cos(x) - 6

# se deriva f(x) para el inciso a
def dfa(x):
    return math.exp(x) - math.log(2)*2**(-x) - 2*math.sin(x)

# se define el inciso b: f(x) = 2x cos(2x) − (x − 2)^2
def fb(x):
    return 2*x*math.cos(2*x) - (x - 2)**2

# se deriva  f(x) para el inciso b
def dfb(x):
    return 2*math.cos(2*x) - 4*x*math.sin(2*x) - 2*(x - 2)

# se define el inciso c: f(x) = e^x − 3x^2
def fc(x):
    return math.exp(x) - 3*x**2

# se deriva  f(x) para el inciso c
def dfc(x):
    return math.exp(x) - 6*x

# se define la funcion que mostrara un menu que permitira elegir el inciso
def menu():
    while True:
        # se imprimen las opciones
        print("\nMetodo de Newton")
        print("Selecciona el inciso deseado:")
        print("a) f(x) = e^x + 2^(-x) + 2cos(x) - 6     en [1, 2]")
        print("b) f(x) = 2x cos(2x) − (x − 2)^2         en [2, 3] y [3, 4]")
        print("c) f(x) = e^x − 3x^2                     en [0, 1] y [3, 5]")
        print("d) Salir")

        opcion = input("Opción: ").lower()  # se lee la opcion dada por el usuario
        # si es a se imprime la solucion
        if opcion == 'a':
            print("\nInciso a) Intervalo [1, 2]")
            sol_a = newton(fa, dfa, x0=1.5)  # se usa 1.5 como estimacion inicial dentro del intervalo
            if sol_a is not None:
                print(f"Raíz: {sol_a:.5f}")
        #si es b se imprime la solucion
        elif opcion == 'b':
            # para la pimera raíz en el intervalo [2, 3]
            print("\nInciso b) Intervalo [2, 3]")
            sol_b1 = newton(fb, dfb, x0=2.5)
            if sol_b1 is not None:
                print(f"Raíz 1: {sol_b1:.5f}")

            # para la segunda raíz en el intervalo [3, 4]
            print("\nInciso b) Intervalo [3, 4]")
            sol_b2 = newton(fb, dfb, x0=3.5)
            if sol_b2 is not None:
                print(f"Raíz 2: {sol_b2:.5f}")
        #si es c se imprime la solucion
        elif opcion == 'c':
            # para la primera raíz en el intervalo [0, 1]
            print("\nInciso c) Intervalo [0, 1]")
            sol_c1 = newton(fc, dfc, x0=0.5)
            if sol_c1 is not None:
                print(f"Raíz 1: {sol_c1:.5f}")

            # para la segunda raíz en el intervalo[3, 5]
            print("\nInciso c) Intervalo [3, 5]")
            sol_c2 = newton(fc, dfc, x0=4.0)
            if sol_c2 is not None:
                print(f"Raíz 2: {sol_c2:.5f}")
        # para salir o finalizar el programa se define
        elif opcion == 'd':
            print("Se sale del programa.")
            break  # se sale del bucle y termina el programa

        else:
            print("la opcion insertada no es valida. Intente de nuevo.")  # se imprime cuando la opción dada por el usuario es invalida

menu()  # se llama al menu
```
