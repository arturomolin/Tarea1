```python
#Ejercicio 6
import numpy as np
import matplotlib.pyplot as plt

# Definir la función g(x) que se nos dan en el problema.
def g(x):
    return np.pi + 0.5 * np.sin(x / 2)

# Hacemos los parámetros de iteración con Valor inicial dentro de [0, 2π]
x0 = 2
tolerancia = 1e-2
max_iter = 100

# Iteración de Punto Fijo
iteraciones = [x0]
for i in range(max_iter):
    x1 = g(x0)
    iteraciones.append(x1)
    if abs(x1 - x0) < tolerancia:
        break
    x0 = x1

# Imprimimos los resultado obtenidos.
print("Iteraciones:")
for i, x in enumerate(iteraciones):
    print(f"Iteración {i}: x = {x:.5f}")
print(f"\nAproximación del punto fijo con tolerancia {tolerancia}: {x1:.5f}")

# Aquí solo ponemos los datos de las graficas.
x_vals = np.linspace(0, 2 * np.pi, 400)
y_vals = g(x_vals)

plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label=r"$g(x) = \pi + 0.5 \sin(x/2)$", color="blue")
plt.plot(x_vals, x_vals, label=r"$y = x$", color="red", linestyle="--")
plt.scatter(iteraciones, [g(x) for x in iteraciones], color='green', label="Iteraciones")
plt.title("Método del Punto Fijo para $g(x)$")
plt.xlabel("x")
plt.ylabel("g(x)")
plt.legend()
plt.grid(True)
plt.show()
```
